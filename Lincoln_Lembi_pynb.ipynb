4) Crie um dicionário de com cinco chaves, em que cada chave é uma fruta e os valores da chaves são palavras que representam o sabor, cor, tamanho, e uma variável booleana indicando se tem semente mais de uma semente na fruta. Por exemplo, Se abacate for uma chave, este pode conter os valores: amargo, verde, médio, False.

Por fim, mostre:

a) A terceira fruta

b) O quarto sabor

c) A segunda cor

d) A primira verificação se há mais de uma semente


```python
#resposta
fruit_dict = {'Maçã':['doce', 'vermelha', 'pequena','True'],
              'Uva':['azeda', 'roxa', 'pequena','true'],
              'melancia':['doce', 'verde', 'grande','false'],
              'banana':['doce', 'amarela', 'media','false'],
              'laranja':['citrica', 'verde', 'media','true']}

# item a)
fruta_terceira = list(fruit_dict.keys())[2]
print(fruta_terceira)
# item b)
sabor_quarto = list(fruit_dict.values())[3][0]
print(sabor_quarto)
#item c)
cor_segunda = list(fruit_dict.values())[1][1]
print(cor_segunda)
#item d)
verificacao_primeira = list(fruit_dict.values())[0][3]
print(verificacao_primeira)
```
